## Probability Distribution and PDF Fitting

### Objective
To generate synthetic data and fit a Gaussian Probability Density Function (PDF) to verify distribution assumptions.

### Theory
A Probability Density Function (PDF) represents the likelihood of a continuous random variable. 
Many Machine Learning algorithms assume Gaussian (Normal) distribution for numerical stability and better performance.

This experiment generates synthetic Gaussian data and verifies the distribution using histogram and PDF fitting.


```python
# ==========================================================
# UCS654 - Experiment 1
# Probability Distribution and PDF Fitting
# ==========================================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis

# -----------------------------
# Step 1: Generate Synthetic Data
# -----------------------------
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# -----------------------------
# Step 2: Compute Statistics
# -----------------------------
mean = np.mean(data)
std_dev = np.std(data)
skewness = skew(data)
kurt = kurtosis(data)

print("Number of Samples:", len(data))
print("Mean (μ):", round(mean, 2))
print("Standard Deviation (σ):", round(std_dev, 2))
print("Skewness:", round(skewness, 4))
print("Kurtosis:", round(kurt, 4))

# -----------------------------
# Step 3: Plot Histogram
# -----------------------------
plt.figure(figsize=(8,5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label="Histogram")

# -----------------------------
# Step 4: Fit Gaussian PDF
# -----------------------------
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, mean, std_dev)

plt.plot(x, pdf, 'r', linewidth=2, label="Fitted Gaussian PDF")

plt.title("Gaussian PDF Fitting")
plt.xlabel("Data Values")
plt.ylabel("Density")
plt.legend()
plt.grid(True)

plt.show()
```

## Observation Table

| Parameter | Observation |
|------------|-------------|
| Number of Samples | 1000 |
| Mean (μ) | ≈ 50 |
| Standard Deviation (σ) | ≈ 10 |
| Skewness | ≈ 0 |
| Kurtosis | ≈ 0 |
| Shape | Bell-shaped |
| Distribution | Approximately Gaussian |

## Conclusion

The histogram closely aligns with the fitted Gaussian PDF.  
The skewness and kurtosis values are near zero, confirming that the dataset approximately follows a normal distribution.

This validates the assumption of normally distributed features in Machine Learning models.
